```python
import pandas as pd

# Get path
WARNING_FOLDER_PATH = "../scraper/tmp/warnings_reviews/20260530_2213.csv"
```

```python
df_w = pd.read_csv(WARNING_FOLDER_PATH, sep="|")
df_w[["episode_name", "warning_type"]].drop_duplicates()["warning_type"].value_counts()
```

### missing_reviewers_info_for_rating    

This episodes don't have info about the reviewers that gave each rating, so we can't make a recommender system with these ones

```python
ep_missing_reviews_info = df_w[
    df_w.warning_type == "missing_reviews_info"
].episode_name.unique()
```

### request_error_rating_url & request_error_url

```python
df_w[
    df_w.warning_type.str.startswith("request_error")
    & (~df_w.episode_name.isin(ep_missing_reviews_info))
][["episode_name", "warning_type"]].drop_duplicates()
```
